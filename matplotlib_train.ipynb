1. Overview of Matplotlib
Brief Introduction:
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It provides a wide variety of plots and customization options, making it a powerful tool for data visualization in various fields including data science, machine learning, and scientific computing.

Why Data Visualization is Important:
* Data visualization helps in understanding complex datasets by presenting information in a graphical format.
* It enables us to identify patterns, trends, and relationships in the data more effectively.
* Visualizations aid in communicating insights and findings to stakeholders in a clear and concise manner.

```python
import matplotlib.pyplot as plt
```

 Line Plot
Introduction:
A line plot is a type of plot that displays data points connected by straight line segments. It's commonly used to visualize the trend of a dataset over a continuous interval.

Using the plot() Function:
In Matplotlib, the plot() function is used to create line plots. It takes two main arguments: the x-coordinates and the y-coordinates of the data points.

```python
import numpy as np
x=np.arange(0,10,0.2)
y=np.sin(x)
# creating a line plot
plt.plot(x,y)
# showing the line plot
plt.show()
```

we can customize various aspects of line plot such as line color,style,marker

* We specify the color of the line using the color parameter.
* We set the line style to dashed using the linestyle parameter.
* We add circular markers at data points using the marker parameter.
* We add labels to x and y axes using xlabel() and ylabel() functions, respectively.
* We set the title of the plot using title() function.
* We add a legend to the plot using legend() function.
* We enable grid lines using grid(True).

```python
plt.plot(x,y,color="red",linestyle="--",marker="o",label="sin(x)")
plt.xlabel("x")
plt.ylabel('sin(x)')
plt.title('Sine function')
plt.legend()
plt.grid(True)
plt.show()
```

 Scatter Plot
Introduction:

A scatter plot is used to visualize the relationship between two variables by displaying data points on a two-dimensional plane. Each point represents an observation in the dataset, with the x-coordinate corresponding to one variable and the y-coordinate corresponding to another.

Using the scatter() Function:
In Matplotlib, the scatter() function is utilized to create scatter plots. It takes the x-coordinates and y-coordinates of the data points as its main arguments.

```python
x=np.random.rand(100)
y=np.random.rand(100)
plt.scatter(x,y)
plt.show()
```

* We specify the color of the markers using the color parameter.
* We set the marker style to triangle (^) using the marker parameter.
* We adjust the size of markers using the s parameter.
* We control the transparency of markers using the alpha parameter.
* Labels, title, legend, and grid lines are added similarly as in line plots.

```python
plt.scatter(x, y, color='blue', marker='^', s=50, alpha=0.5, label='Random Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Generate random data from a normal distribution
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

# Display the plot
plt.show()
```

```python
categories=["A","B","C","D"]
data=[10,20,30,40]
plt.bar(categories,data,color="red")
plt.xlabel("categories")
plt.ylabel("data")
plt.title("Bar plot")
plt.show()
```

```python
plt.pie(data,labels=categories,autopct="%1.1f%%",startangle=90)
plt.axis('equal')
plt.show()
```

Introduction:
Subplots allow you to create multiple plots within the same figure. They are useful for comparing different datasets or visualizing related information side by side.

Using the subplot() Function:
In Matplotlib, the subplot() function is used to create subplots. It takes three main arguments: the number of rows, the number of columns, and the index of the subplot to create.

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)
plt.subplot(1,2,1)
plt.plot(x,y1,color="red")
plt.title("sin(x)")
plt.subplot(1,2,2)
plt.plot(x,y2,color="green")
plt.title("cos(x)")
plt.tight_layout()
plt.savefig("outputsinandcos.jpg")
plt.show()
```
